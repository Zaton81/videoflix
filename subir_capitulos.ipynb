### Juego de Tronos

```python
#recogemos las urls

url1 = "https://es.wikipedia.org/wiki/Anexo:Primera_temporada_de_Game_of_Thrones"
url2 = "https://es.wikipedia.org/wiki/Anexo:Segunda_temporada_de_Game_of_Thrones"
url3 = "https://es.wikipedia.org/wiki/Anexo:Tercera_temporada_de_Game_of_Thrones"
url4 = "https://es.wikipedia.org/wiki/Anexo:Cuarta_temporada_de_Game_of_Thrones"
url5 = "https://es.wikipedia.org/wiki/Anexo:Quinta_temporada_de_Game_of_Thrones"
url6 = "https://es.wikipedia.org/wiki/Anexo:Sexta_temporada_de_Game_of_Thrones"
url7 = "https://es.wikipedia.org/wiki/Anexo:S%C3%A9ptima_temporada_de_Game_of_Thrones"
url8 = "https://es.wikipedia.org/wiki/Anexo:Octava_temporada_de_Game_of_Thrones"
urls=[url1, url2, url3, url4, url5, url6, url7, url8]
```

```python
#almacenamos los datos

data = []

for url in urls:
    data.append(requests.get(url).text)
data
```

```python
#parseamos los datos y añadimos al dataframe
#soup = BeautifulSoup(data, 'html.parser')
list_soup = [] 
for dato in data:
    list_soup.append(BeautifulSoup(dato, "html.parser"))
```

```python
#parseamos los datos


# Crear una lista para almacenar los DataFrames
dataframes_list = []

# Iterar sobre las tablas (suponiendo que hay 8 tablas)
for soup in list_soup:
    # Crear el DataFrame
    juego_df = pd.DataFrame(columns=["N en serie", "Ntemporada", "Título", "Drigidopor", "Escrito", 
                                     "Fecha", "audiencia (millones)", "sinopsis"])

    # Inicializar variables para almacenar datos temporales
    nserie, ntemporada, titulo, dirigido, escrito, fecha, audiencia = None, None, None, None, None, None, None

    # Iterar sobre las filas de la tabla
    for row in soup.find_all("tbody")[2].find_all("tr"):
        if 'vevent' in row.get('class', []):
            # Filas de episodios
            col = row.find_all("td")
            nserie = row.find("th").text.strip()
            ntemporada = col[0].text.strip()
            titulo = col[1].text.strip()
            dirigido = col[2].text.strip()
            escrito = col[3].text.strip()
            fecha = col[4].text.strip()
            audiencia = col[5].text.strip()
        elif 'expand-child' in row.get('class', []):
            # Filas de sinopsis
            sinopsis = row.find("td", class_="description").text.strip()
            
            # Añadir la fila al DataFrame
            juego_df = pd.concat([juego_df, pd.DataFrame({"N en serie": [nserie], "Ntemporada": [ntemporada], "Título": [titulo], 
                                                          "Drigidopor": [dirigido], "Escrito": [escrito], "Fecha": [fecha], 
                                                          "audiencia (millones)": [audiencia], "sinopsis": [sinopsis]})], ignore_index=True)
    
    # Añadir el DataFrame a la lista
    dataframes_list.append(juego_df)

# Mostrar los DataFrames
for df in dataframes_list:
    print(df)
```
